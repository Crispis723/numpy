##2 Datos numericos

```python
columnas=list(datos.columns)
columnas
```

```python
datos=datos.explode(columnas[3:])#explota cada elemento de una variable saca los elementos osea cea registros apartir de el registro encapsulado
datos
```

```python
datos.reset_index(drop=True,inplace=True) #implace para el mismo dataframe y drop borar indice
datos.head(5)
```

```python
datos.info()
```

```python
import numpy as np
```

```python
datos["max_hospedes"]=datos["max_hospedes"].astype(np.int64) #para cambiar el tipo de datos
```

```python
datos.info()
```

```python
columnas_numericas=["cantidad_baños","cantidad_cuartos","cantidad_camas"]
```

```python
datos[columnas_numericas]=datos[columnas_numericas].astype(np.int64)
```

```python
datos["evaluacion_general"]=datos["evaluacion_general"].astype(np.float64)
```

```python
datos["precio"]=datos["precio"].apply(lambda x:x.replace("$","").replace(",","").strip())
#lambda para todos los valores de una columna, replace es remplazar,strip quita espacios y apply  solo funciona con una sola columna(series)
datos["precio"]
```

```python
datos["precio"]=datos["precio"].astype(np.float64)
```

```python
datos.info()
```

```python
datos[["cuota_deposito","cuota_limpieza"]]=datos[["cuota_deposito","cuota_limpieza"]].applymap(lambda x:x.replace("$","").replace(",","").strip()) #applymap con dataframe
```

```python
datos[["cuota_deposito","cuota_limpieza"]].astype(np.float64)
```

```python
datos["descripcion_local"]=datos["descripcion_local"].str.lower()
```

```python
datos.head()
```

```python
datos["descripcion_local"][3169]
```

```python
datos["descripcion_local"]=datos["descripcion_local"].str.replace("[^a-zA-Z0-9\-\s]"," ",regex=True) # regex se utiliza para remplazar muchos caracteres
```

```python
datos["descripcion_local"]=datos["descripcion_local"].str.replace("(?<!\w)-(?!\w)"," ",regex=True)
```

```python
datos["descripcion_local"].head()
```

```python
datos["descripcion_local"]=datos["descripcion_local"].str.split() # split  genera una lista usando un separado  y convertida en lista  llamada tokens
datos.head()
```

```python
datos["comodidades"]=datos["comodidades"].str.replace('\{|}\"', '', regex=True)
```

```python
datos.head()
```

```python
dt_data=pd.read_json("inmuebles_disponibles.json")
dt_data.head()
```

```python
dt_data.info()
```

```python
dt_data["fecha"]=pd.to_datetime(dt_data["fecha"])
```

```python
dt_data.info()
```

```python
dt_data.head()
```

```python
dt_data["fecha"].dt.strftime("%Y-%m")#año-mes-dia
```

```python
subset=dt_data.groupby(dt_data["fecha"].dt.strftime("%Y-%m"))["lugar_disponible"].sum() #solo funciona con boleanos solo funciona  con valores true
subset
```

```python

# utilizamos el método fillna para llenar los elementos vacíos por '0.0'
# definimos el parámetro de inplace para True para substituir en el DataFrame
dt_data['precio'].fillna('0.0', inplace = True)

# borramos el $ y las comas con apply lambda
dt_data['precio'] = dt_data['precio'].apply(lambda x: x.replace('$', '').replace(',',''))

# transformamos los tipos de datos para float64
dt_data['precio'] = dt_data['precio'].astype(np.float64)

# observamos el resultado final
dt_data
```
