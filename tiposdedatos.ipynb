#Aula 5 - Leyendo banco de datos

##Creando un banco local de datos

```python
import sqlalchemy #biblioteca para manejar sql
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect,text # create rueda la base de datos, visionario de todos los datos
```

```python
engine = create_engine("sqlite:///:memory:")# ///=memoria local
```

##*Escribiendo en un banco de datos

```python
sql="clientes_banco.csv"
```

```python
clientes_banco=pd.read_csv(sql)
```

```python
clientes_banco.head(4)
```

```python
clientes_banco.to_sql("clientes", engine, index=False)
```

```python
inspect=inspect(engine)
```

```python
print(inspect.get_table_names())
```

##Leyendo una consulta SQL

```python
query= "SELECT * FROM clientes WHERE Categoria_de_renta = 'Empleado'"
```

```python
Empleados=pd.read_sql(sql=text(query), con=engine.connect())
```

```python
Empleados
```

```python
Empleados.to_sql("empleados", con=engine.connect(), index=False)
```

```python
pd.read_sql_table("empleados", con=engine.connect(), columns=["ID_Cliente", "Grado_estudio", "Rendimiento_anual"])
```

```python
query= "SELECT * FROM clientes"
```

```python
pd.read_sql(sql=text(query), con=engine.connect())
```

```python
#Borrando registro
from sqlalchemy.exc import SQLAlchemyError
query="DELETE FROM clientes WHERE ID_Cliente=5079028"
try:
  r_set=engine.connect().execute(text(query))
#engine conect conecta al la base  , execute ejecuta
#query es el codigo pero en base
except SQLAlchemyError as e:
  print(e)
else:
  print("Registro borrados", r_set.rowcount)
```

```python
pd.read_sql_table("clientes", con=engine.connect())
```

```python
#actualizar registro
query="UPDATE clientes SET 	Grado_estudio ='Nivel superior' WHERE ID_Cliente=5079030	"
try:
  r_set=engine.connect().execute(text(query))
#engine conect conecta al la base  , execute ejecuta
#query es el codigo pero en base
except SQLAlchemyError as e:
  print(e)
else:
  print("Registro actualizados", r_set.rowcount)
```

```python
pd.read_sql_table("clientes", con=engine.connect())
```

```python
inspect=inspect(engine)
print(inspect.get_table_names())
```
